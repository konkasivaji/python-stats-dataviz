```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape
```

```python
df.sample(5)
```

```python
df["indicator"].unique()
```

```python
df1 = df.query('indicator == "SP.POP.TOTL"')    #Taking population
df1
```

```python
df["Country Name"].unique()
```

```python
my_countries = ["Albania", "Vietnam", "Switzerland","United Kingdom","Pakistan"]

df2 = df1[df1["Country Name"].isin(my_countries)]

df2.shape
```

```python
df3 = df2.query(' Year == 2021')

df3
```

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021  Population rate of the listed countries",
    color="Country Name", 
    template="plotly_white"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 Population of the countries", 
    color="Country Name", 
    template="plotly_white"
)

fig.update_layout(showlegend=True)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
fig = px.box(df3, x="Country Name", y="value")
fig.show()
```

```python
fig = px.bar(
    df3, 
    x="Region", 
    y="value", 
    title="2021 Population of Countries ",
    color="Country Name", 
    template="seaborn"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Region", 
    values="value", 
    title="2021 Polpulation of selected regions", 
    color="Country Name", 
    template="seaborn"
)

fig.update_layout(showlegend=True)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
df4=df3.groupby('Region')['value'].mean()
df4
```

```python
fig = px.bar(x=df4.index, y=df4)
fig.show()
```

```python
fig = px.pie(df4,names=df4.index,
    values="value", 
    title="2021 Population of selected regions",  
    template="plotly_white"
)
fig.show()
```

```python
df5 = df2.groupby(['Region', 'Year']).agg({'value': ['mean']})
df5.columns = ['value_mean']
df5 = df5.reset_index()
print(df5)
```

```python
fig = px.line(
    df5,
    x="Year",
    y="value_mean", 
    title="2021 Population of the Selected Countries", 
    color="Region", 
    template="plotly_white"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
df6 = df.query(' Year == 2020')
df6
```

```python
my_indicators = ["SP.DYN.LE00.IN", "SP.POP.TOTL"]

df7 = df6[df6["indicator"].isin(my_indicators)]

df7.shape
```

```python
df8 = pd.pivot_table(df7, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df8.head()
```

```python
df8.dropna(inplace=True)

df8.head()
```

```python
fig = px.scatter(
    df8,
    x="SP.POP.TOTL",
    y="SP.DYN.LE00.IN", 
    title="Population& Health", 
    color="Country Name", 
    template="plotly_white"
)

fig.update_layout(showlegend=True)

fig.show()
```
